# 2. Analyze trip durations and distances.

```python
#convert the columns involve to date and time 
Taxi_Trip_2018['lpep_pickup_datetime'] = pd.to_datetime(Taxi_Trip_2018['lpep_pickup_datetime'])
Taxi_Trip_2018['lpep_dropoff_datetime'] = pd.to_datetime(Taxi_Trip_2018['lpep_dropoff_datetime'])
```

```python
# Calculate trip duration in minutes
Taxi_Trip_2018['trip_duration'] = (Taxi_Trip_2018['lpep_dropoff_datetime'] - Taxi_Trip_2018['lpep_pickup_datetime']).dt.total_seconds() / 60
```

```python
# Remove negative or very short durations (possible data errors less than 0)
Taxi_Trip_2018 = Taxi_Trip_2018[Taxi_Trip_2018['trip_duration'] > 0]
```

```python
# Summary statistics for trip durations and distances
duration_stats_2018 = Taxi_Trip_2018['trip_duration'].describe()
distance_stats_2018 = Taxi_Trip_2018['trip_distance'].describe()
```

```python
print("\nTrip Duration Statistics (2018):")
print(duration_stats_2018)
print("\nTrip Distance Statistics (2018):")
print(distance_stats_2018)
```

```python
# Visualize the distribution of trip durations and distances using histograms
plt.figure(figsize=(14, 6))

# Histogram for trip duration (2018)
plt.subplot(2, 2, 3)
plt.hist(Taxi_Trip_2018['trip_duration'], bins=50, color='blue', edgecolor='black')
plt.title('Trip Duration Distribution (2018)')
plt.xlabel('Trip Duration (minutes)')
plt.ylabel('Frequency')

# Histogram for trip distance (2018)
plt.subplot(2, 2, 4)
plt.hist(Taxi_Trip_2018['trip_distance'], bins=50, color='green', edgecolor='black')
plt.title('Trip Distance Distribution (2018)')
plt.xlabel('Trip Distance (miles)')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

# Visualize using box plots to identify outliers
plt.figure(figsize=(14, 6))

# Box plot for trip distance
plt.subplot(1, 2, 2)
plt.boxplot([Taxi_Trip_2018['trip_distance'], Taxi_Trip_2018['trip_distance']], labels=['2017', '2018'])
plt.title('Trip Distance Boxplot')
plt.ylabel('Trip Distance (miles)')

plt.tight_layout()
plt.show()
```

# Identify any seasonal fare trends.

```python
Taxi_Trip_2018['month'] = Taxi_Trip_2018['lpep_pickup_datetime'].dt.month
```

```python
## Define seasons based on months
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    elif month in [9, 10, 11]:
        return 'Fall'
```

```python
Taxi_Trip_2018['season'] = Taxi_Trip_2018['month'].apply(get_season)
```

```python
# Group by season and calculate the average fare
seasonal_fare_2018 = Taxi_Trip_2018.groupby('season')['fare_amount'].mean()

# Print seasonal fare trends
print("\nSeasonal Fare Trends (2018):\n", seasonal_fare_2018)
```

```python
# Visualize Seasonal Fare Trends using a bar plot
plt.figure(figsize=(10, 6))

# Create a DataFrame to store seasonal fare trends for visualization
seasonal_fares_df = pd.DataFrame({'2018': seasonal_fare_2018})

# Plot the seasonal fare trends
seasonal_fares_df.plot(kind='bar', color=['blue', 'green'], figsize=(10, 6))

# Add titles and labels
plt.title('Average Fare Amount by Season (2017 vs 2018)', fontsize=14)
plt.xlabel('Season', fontsize=12)
plt.ylabel('Average Fare (USD)', fontsize=12)

# Show the plot
plt.tight_layout()
plt.grid(True)
plt.show()

# Optional: Visualize Monthly Trends
plt.figure(figsize=(10, 6))

# Group by month and calculate the average fare
monthly_fare_2018 = Taxi_Trip_2018.groupby('month')['fare_amount'].mean()

# Line plot for monthly average fares
plt.plot(monthly_fare_2018.index, monthly_fare_2018.values, label='2018', color='green', marker='o')

# Add titles and labels
plt.title('Average Fare Amount by Month (2018)', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Average Fare (USD)', fontsize=12)
plt.xticks(ticks=range(1, 13), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.legend()

# Show the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```
